```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters for the circle
angle = np.linspace(0, 2 * np.pi, 150)
radius = 1

# Cartesian coordinates of the circle
x = radius * np.cos(angle)
y = radius * np.sin(angle)

# Plotting the circle
plt.figure(figsize=(6,6))
plt.plot(x, y)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.title('Unit Circle')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')  # Set equal scaling by changing aspect ratio of the plot box
plt.show()
```

```python
# Parameters for the circle
angle = np.linspace(0, 2 * np.pi, 150)
radius = 1

# Cartesian coordinates of the circle
x = radius * np.cos(angle)
y = radius * np.sin(angle)

# Create a mask for the segments of the circle where x < -1/2 or x > 1/2
red_mask = (x < -0.5) | (x > 0.5)

# Cartesian coordinates for the red and the rest of the circle points
x_red = x[red_mask]
y_red = y[red_mask]
x_rest = x[~red_mask]
y_rest = y[~red_mask]

# Create a figure and axis with equal aspect ratio
fig, ax = plt.subplots(figsize=(6,6))

# Plot the rest of the circle points as blue dots
ax.plot(x_rest, y_rest, 'o', color='blue', label='Remaining Circle Points')  # Blue dots
# Plot the red points as red dots
ax.plot(x_red, y_red, 'o', color='red', label='X < -1/2 or X > 1/2')  # Red dots

# Set the limits, labels, title and grid for the plot
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_title('Unit Circle with Specific Points Colored')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.grid(True)
# Draw a diameter of the circle in black
ax.plot([-1, 1], [0, 0], color='black', label='Diameter')  # Black line

angle1 = -np.arccos(-0.5)
angle2 = -np.arccos(0.5)

ax.plot([0, -np.cos(angle1)], [0, np.sin(angle1)], color='green')
ax.plot([0, -np.cos(2*np.pi - angle1)], [0, np.sin(2*np.pi - angle1)], color='green')

ax.legend()

# Show the plot
plt.show()
```
