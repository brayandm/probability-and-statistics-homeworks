```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function for the first segment
def f_rho_segment1(r):
    return 2 * r

# Create ranges for each segment
r_segment1 = np.linspace(0, 1, 100)  # Segment from 0 to 1
r_segment2 = np.linspace(-1, 0, 50)  # Segment less than 0
r_segment3 = np.linspace(1, 2, 50)   # Segment greater than 1

# Plot each segment
plt.plot(r_segment1, f_rho_segment1(r_segment1), label=r'$2r \; (0 \leq r \leq 1)$')
plt.plot(r_segment2, np.zeros_like(r_segment2), label=r'$0 \; (r < 0)$')  # Zero array for y values
plt.plot(r_segment3, np.zeros_like(r_segment3), label=r'$0 \; (r > 1)$')  # Zero array for y values

# Add labels and title
plt.xlabel('r')
plt.ylabel(r'$f_\rho(r)$')
plt.title('Graph of the Function $f_\\rho(r)$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function for the first segment
def f_phi_segment1(theta):
    return np.full_like(theta, 1 / np.pi)

# Create ranges for each segment
theta_segment1 = np.linspace(0, np.pi, 100)  # Segment from 0 to pi
theta_segment2 = np.linspace(-1, 0, 50)      # Segment less than 0
theta_segment3 = np.linspace(np.pi, np.pi + 1, 50)  # Segment greater than pi

# Plot each segment
plt.plot(theta_segment1, f_phi_segment1(theta_segment1), label=r'$\frac{1}{\pi} \; (0 \leq \theta \leq \pi)$')
plt.plot(theta_segment2, np.zeros_like(theta_segment2), label=r'$0 \; (\theta < 0)$')
plt.plot(theta_segment3, np.zeros_like(theta_segment3), label=r'$0 \; (\theta > \pi)$')

# Add labels and title
plt.xlabel(r'$\theta$')
plt.ylabel(r'$f_\phi(\theta)$')
plt.title('Graph of the Function $f_\\phi(\\theta)$')
plt.legend()
plt.grid(True)
plt.show()
```
